```python
import pandas as pd
```

```python
abspath = '/Users/spencerbertsch/Desktop/winequality-red.csv'
df = pd.read_csv(abspath, sep=';')
```

```python
[col for col in df.columns]
```

```python
predictors: list = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide',
 'density', 'pH', 'sulphates', 'alcohol']
response: list = ['quality']

# define our features and labels
X = df[predictors]
y = df[response]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=23)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
    max_depth=5, random_state=0).fit(X_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train) # Note: x_0 =1 is no need to add, sklearn will take care of it.
```

```python
pd.DataFrame(lin_reg.coef_).T
```

```python
list(lin_reg.intercept_) + list(lin_reg.coef_)
```

```python
Beta = list(lin_reg.coef_[0])
```

```python
# sklearn regression module
y_pred = lin_reg.predict(X_test)

#Evaluvation: MSE
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred, y_test)

# R_square
R_square_sk = lin_reg.score(X_test, y_test)
print('The Mean Square Error(MSE) or J(theta) is: ',mse)
print('R square obtain for scikit learn library is :',R_square_sk)
```
